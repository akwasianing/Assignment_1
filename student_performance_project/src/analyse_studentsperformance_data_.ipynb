```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# load cleaned_studentsperformance_data
df = pd.read_csv('https://raw.githubusercontent.com/akwasianing/Assignment_1/refs/heads/main/student_performance_project/data_clean/cleaned_studentsperformance_data.csv')
df.head(10)
```

```python
# First visualization
# Math Score Distribution by Parental Education Level using boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x="parental level of education", y="math score", data=df, hue="parental level of education", palette="Set2")
plt.xticks(rotation=45)
plt.title("Math Score Distribution by Parental Education Level")
plt.show()
```

```python
# Second visualization
# Count of Test Preparation Course Completion by Gender
plt.figure(figsize=(6,5))
sns.countplot(x='test preparation course', hue='gender', data=df, palette='Set2')
plt.title("Count of Test Preparation Course Completion by Gender")
plt.xlabel("Test Preparation Course")
plt.ylabel("Count")
plt.show()
```

```python
# Third visualization
# Scatter plot to visualize the relationship between math scores and reading scores, with an additional color coding based on gender.
sns.scatterplot(x='math score', y='reading score', hue='gender', data=df)
plt.title("Scatter Plot of Math vs. Reading Scores, Colored by Gender")
plt.show()
```

```python
# Fourth visualization
# Boxplot to display the impact of test preparation on writing scores
sns.boxplot(x='test preparation course', y='writing score', data=df)
plt.title("Impact of Test Preparation on Writing Scores")
plt.show()
```

```python
# Fifth visualization
# Boxplot to displah math scores by parental education level and test preparation course
plt.figure(figsize=(10, 6))
sns.boxplot(x="parental level of education", y="math score", hue="test preparation course", data=df, palette="Set2")
plt.title("Math Scores by Parental Education Level and Test Preparation Course")
plt.xlabel("Parental Level of Education")
plt.ylabel("Math Score")
plt.xticks(rotation=45)
plt.legend(title="Test Preparation Course", loc="upper right")
plt.show()
```
