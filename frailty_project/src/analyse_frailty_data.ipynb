```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns




```

```python
# load cleaned_frailty_data
df = pd.read_csv('https://raw.githubusercontent.com/akwasianing/Assignment_1/refs/heads/main/frailty_project/data_clean/cleaned_frailty_data.csv')
df
```

```python
# Display Basic Statistics
summary = df.describe()
print(summary)

# Save summary statistics
summary.to_csv("result.txt")
```

```python
# Create a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Frailty', data=df)
plt.title('Correlation between Age and Frailty')
plt.xlabel('Age')
plt.ylabel('Frailty')
plt.show()
```
