```python
# function to identify outliers based on the IQR method
def grap_outliers(df):
  q1 = df.quantile(0.25)
  q3 = df.quantile(0.75)
  iqr = q3 - q1
  outliers = df[((df<(q1-1.5*iqr)) | (df>(q3+iqr*1.5)))]
  return outliers

for column in df[['Height', 'Weight', 'Age', 'Grip strength']]:
  outliers = grap_outliers(df[column])
  print(f'Number of outliers in {column}: {len(outliers)}')
```
